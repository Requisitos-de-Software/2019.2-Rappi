```python
import pandas as pd
import numpy
```

```python
columns = []
index = [n for n in range(1, 25)]
df = pd.DataFrame(index=index, columns=columns)

beneficio_relativo = [3,5,6,4,8,7,8,7,4,4,6,7,4,3,7,8,7,5,2,2,2,3,3,8]
penalidade_relativa = [9,9,9,7,3,4,5,8,2,6,8,6,5,3,6,2,7,4,5,1,5,6,2,1]
df['Benefício Relativo'] = beneficio_relativo
df['Penalidade Relativa'] = penalidade_relativa

df = df.assign(valor_total = lambda x: x['Benefício Relativo']*1.0 + x['Penalidade Relativa']*1.5)
df = df.rename({'valor_total': 'Valor Total'}, axis=1)

df = df = df.assign(valor_percent = lambda x: x['Valor Total']*100/df['Valor Total'].sum())
df = df.rename({'valor_percent': 'Valor(%)'}, axis=1)

custo_relativo = [4,4,6,7,5,5,8,7,2,3,4,5,4,5,6,4,7,6,4,6,4,6,2,7]
df['Custo Relativo'] = custo_relativo

df = df = df.assign(custo_percent = lambda x: x['Custo Relativo']*100/df['Custo Relativo'].sum())
df = df.rename({'custo_percent': 'Custo(%)'}, axis=1)

risco_relativo = [2,2,9,7,6,5,9,7,3,2,5,6,4,3,7,2,7,7,3,5,6,5,3,8]
df['Risco Relativo'] = risco_relativo

df = df = df.assign(risco_percent = lambda x: x['Risco Relativo']*100/df['Risco Relativo'].sum())
df = df.rename({'risco_percent': 'Risco(%)'}, axis=1)

df = df = df.assign(prioridade = lambda x: x['Valor(%)']/(x['Custo(%)'] + x['Risco(%)']*1.2))
df = df.rename({'prioridade': 'Prioridade'}, axis=1)

df = df.sort_values(by='Prioridade', ascending=False)
df = df.replace('.', ',')

df
```

```python
df.to_csv('first_things_first.csv')
```
